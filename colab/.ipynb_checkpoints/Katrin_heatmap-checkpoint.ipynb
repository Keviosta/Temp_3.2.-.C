```python
#data Proteomics
df = pandas.read_csv(file_directory + input_dir + 'clustered_sign_genes2.csv', 
                     sep = ',', index_col = 'gene')

print(df.shape, '\n')
df.head(3)
```

## Remove cluster column. do a new one. Find best number cluster. Can do like that or mean of the triplicate 

```python
# too much Clusters
print(df.cluster.max())
df = df.drop(['Unnamed: 0', 'cluster'], axis =1)
df.head()
```

```python
# Convert to Z_score
df_zscore = stats.zscore(df, axis = 1)
df_zscore.head(2)
```

```python
# Calculate Similarity/Distance
from sklearn.cluster import KMeans

# Determine the optimal number of clusters using the elbow method
# "inertia" is the sum of squared distances of samples to their closest cluster center.
inertia = []
for n_clusters in range(1, 10):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init='auto')
    kmeans.fit(df_zscore)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 10), inertia, marker='o')
plt.title('Identify the Elbow Point')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
import scipy.cluster.hierarchy as sch

# Create the dendrogram to determine the optimal clusters
dendrogram = sch.dendrogram(sch.linkage(df_zscore, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Genes')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
## plot heatmap
g =  seaborn.clustermap(df_zscore, 
                        cmap='bwr', 
                        row_cluster=True, 
                        col_cluster=False, 
#                         vmin=-6.5, vmax=6.5,
                        cbar_kws={'label':'Protein abundance\n by TMT reporter ion intensity\n (z-score)'},
                        xticklabels=True, yticklabels=False, 
                        method='complete', metric='cosine')

g.ax_heatmap.xaxis.set_ticks_position('top')
g.ax_heatmap.xaxis.set_label_position('top')
```

```python
# cluster = 6 
```

```python
sample_names = df_zscore.index
num_samples = len(sample_names)

reordered_indices = g.dendrogram_row.reordered_ind #the index in the heatmap
linkage_matrix  = g.dendrogram_row.linkage #the matrix for heatmap
num_clusters = 4 # cluster expected

#Get the cluster labels for each sample
cluster_labels = fcluster(linkage_matrix, num_clusters, criterion='maxclust')
print(cluster_labels[:10])


subgroup_1 = [sample_names[i] for i in range(len(sample_names)) if cluster_labels[i] == 1]
print(f'n = {len(subgroup_1)} in cluster 1')

subgroup_2 = [sample_names[i] for i in range(len(sample_names)) if cluster_labels[i] == 2]
print(f'n = {len(subgroup_2)} in cluster 2')

subgroup_3 = [sample_names[i] for i in range(len(sample_names)) if cluster_labels[i] == 3]
print(f'n = {len(subgroup_3)} in cluster 3')

subgroup_4 = [sample_names[i] for i in range(len(sample_names)) if cluster_labels[i] == 4]
print(f'n = {len(subgroup_4)} in cluster 4')

# subgroup_5 = [sample_names[i] for i in range(len(sample_names)) if cluster_labels[i] == 5]
# print(f'n = {len(subgroup_5)} in cluster 5')
      
# subgroup_6 = [sample_names[i] for i in range(len(sample_names)) if cluster_labels[i] == 6]
# print(f'n = {len(subgroup_6)} in cluster 6')

print(subgroup_1[:10])
```

```python
# See if the selection is good
subgroups = subgroup_1, subgroup_2, subgroup_3, subgroup_4

for subgroup in subgroups:
    plt.subplots(figsize=(4,4))
    
    seaborn.heatmap(df_zscore[df_zscore.index.isin(subgroup)],
                    annot=False, cmap="bwr", 
                    linewidths=0, center=0, vmin=-6.5, vmax=6.5,
                    cbar_kws={'label': 'Zscore \n protein expression'})
    plt.show()
```
