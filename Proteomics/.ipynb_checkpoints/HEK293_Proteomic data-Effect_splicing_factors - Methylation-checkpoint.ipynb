## GSEA: Splicing Factors

```python
# Tag % = % of gene hits before (for positive ES) or after (for negative ES) the peak
#         in the running enrichment score. This gives an indication of the percentage 
#         of genes contributing to the enrichment score.

# lead_genes: leading edge genes (gene hits before running enrichment peak)
```

```python
BP_terms = ['GO:0008380', 'GO:0000398']
CC_terms = ['GO:0005681', 'GO:0005685', 'GO:0005686', 'GO:0005689', 'GO:0071013', 'GO:0071014', 'GO:0030532']
MF_terms = ['GO:0003723']
```

```python
# Rank and so GSEA plot will be based on SLope . More stat is high more it's diff expressed.
df_linregress.reset_index(inplace = True)
ranking = df_linregress[['Gene', 'slope']].dropna().sort_values('slope', ascending = False)
print(ranking.shape)
```

### Cellular Component

```python
# GO_cellular component_2023
pre_res = gseapy.prerank(rnk = ranking.set_index("Gene", drop=True),
                     gene_sets = ['GO_Cellular_Component_2023'],
                     seed = 6, permutation_num = 100)

enrich_results = pre_res.res2d.sort_values("NES", ascending = False)
terms = enrich_results.Term
terms = terms[terms.str.contains('|'.join(CC_terms))]

# positive
print('positive: \n', pre_res.res2d.iloc[:3,[1,7,8]])
pre_res.plot(terms=terms)
plt.show()
plt.close()
```

### Molecular function

```python
terms[terms.str.contains('|'.join(MF_terms))]
```

### Biological Process

```python
# GO_Biological_Process_2023
pre_res = gseapy.prerank(rnk = ranking.set_index("Gene", drop=True),
                     gene_sets = ['GO_Biological_Process_2023'],
                     seed = 6, permutation_num = 100)

enrich_results = pre_res.res2d.sort_values("NES", ascending = False)
terms = enrich_results.Term
terms = terms[terms.str.contains('|'.join(BP_terms))]

# positive
print(pre_res.res2d[pre_res.res2d['Term'].str.contains('|'.join(BP_terms))].iloc[:,[1,7,8]])
pre_res.plot(terms=terms)

plt.show()
plt.close()
```

## GSEA: Methylation 

```python
# Tag % = % of gene hits before (for positive ES) or after (for negative ES) the peak
#         in the running enrichment score. This gives an indication of the percentage 
#         of genes contributing to the enrichment score.

# lead_genes: leading edge genes (gene hits before running enrichment peak)
```

```python
BP_terms = ['GO:0001510', 'GO:0006349', 'GO:0032259', 'GO:0034968', 'GO:0043414']
MF_terms = ['GO:0008173', 'GO:0008968', 'GO:0018024']
```

```python
# Rank and so GSEA plot will be based on SLope . More stat is high more it's diff expressed.
df_linregress.reset_index(inplace = True)
ranking = df_linregress[['Gene', 'slope']].dropna().sort_values('slope', ascending = False)
print(ranking.shape)
```

### Molecular function

```python
terms[terms.str.contains('|'.join(MF_terms))]
```

### Biological Process

```python
# GO_Biological_Process_2023
pre_res = gseapy.prerank(rnk = ranking.set_index("Gene", drop=True),
                     gene_sets = ['GO_Biological_Process_2023'],
                     seed = 6, permutation_num = 100)

enrich_results = pre_res.res2d.sort_values("NES", ascending = False)
terms = enrich_results.Term
terms = terms[terms.str.contains('|'.join(BP_terms))]

# positive
print(pre_res.res2d[pre_res.res2d['Term'].str.contains('|'.join(BP_terms))].iloc[:,[1,7,8]])
pre_res.plot(terms=terms)

plt.show()
plt.close()
```

```python
lead_genes_RNA_methyl = pre_res.res2d.loc[pre_res.res2d["Term"].str.contains('GO:0001510')].Lead_genes
lead_genes_RNA_methyl = lead_genes_RNA_methyl.str.split(';')
rna_methyl_prot = lead_genes_RNA_methyl.explode()
print(len(rna_methyl_prot))

rna_methyl_prot[:20]
```
