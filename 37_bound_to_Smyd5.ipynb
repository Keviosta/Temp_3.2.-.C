# Data Downloads

```python
%%time
#data mouse_NPC
mouse_NPC_df = pandas.read_csv(file_directory+'1) input/mNPC_all_detected_genes_DEseq.csv',
                               sep = ',')
print(mouse_NPC_df.shape)
mouse_NPC_df
```

```python
# download the Smyd5 bound list
with open(file_directory+'1) input/SMYD5_bound_list.txt') as f:
    SMYD5_bound_list = f.read().splitlines()
```

```python
print(len(SMYD5_bound_list))
selected_df = mouse_NPC_df[mouse_NPC_df['gene'].isin(SMYD5_bound_list)].copy()
print(len(selected_df))
selected_df.head()
```

```python
selected_df = selected_df[['gene', 'log2FoldChange', 'pvalue']]
selected_df.reset_index(drop= True, inplace = True)
selected_df.head()
```

```python
heatmap_df = selected_df[['gene','log2FoldChange']]
heatmap_df.set_index('gene', inplace= True)

plt.subplots(figsize=(17,2))
plt.tight_layout()
seaborn.heatmap(heatmap_df.T, annot=False, cmap="bwr", linewidths=.5, center=0, vmin=-1.5, vmax=1.5,
               cbar_kws={'label': 'log2(FC)'})
plt.title('Expression difference between 32 and 37°C \n of the Smyd5-bound genes')
title_file = 'Heatmap_mNPC'
plt.savefig(file_directory+'3) output/'+title_file+'.svg')
```

### 2nd idea. Not good, need to transform in zscore to have pos and neg. To transform, need to have variance stabilization.

```python
selected_df = selected_df.assign(mean_37deg= selected_df[['37_degrees', "37_degrees.1", 
                                                           "37_degrees.2"]].mean(axis=1),
                                 
                                 mean_32deg= (selected_df[['32_degrees', "32_degrees.1", 
                                                           "32_degrees.2", "32_degrees.3"]].mean(axis=1)))
        
                                              
selected_df.head()
```

```python
selected_df = selected_df[['gene', 'mean_37deg', 'mean_32deg']]
selected_df.reset_index(drop= True, inplace = True)
selected_df.head()
```

```python
selected_df = selected_df.assign(FC = selected_df['mean_32deg'] / selected_df['mean_37deg'])
selected_df.head()
```

```python
heatmap_df = selected_df[['gene','FC']]
heatmap_df.set_index('gene', inplace= True)

plt.subplots(figsize=(12,2))
plt.tight_layout()
seaborn.heatmap(heatmap_df.T, annot=False, cmap="bwr", linewidths=.5, center=0, vmin=-1.5, vmax=1.5,
               cbar_kws={'label': 'FC(counts)'})
plt.title('Expression difference between 32 and 37°C \n of the Smyd5-bound genes')
```
