# Data Download

```python
#data DEseq HEK293 significant ones
df_GExp_all = pandas.read_csv(file_directory+'1) input/HEK293_temp_DESeq.csv',
                               sep = ',')
print(df_GExp_all.shape, '\n', df_GExp_all.columns)

# select columns and rename
columns = ['log2FoldChange','padj','gene']
df_GExp_all = df_GExp_all[columns]
df_GExp_all = df_GExp_all.rename(columns = {'log2FoldChange' : 'log2_FC (GEx)',
                                            'padj' : 'padj(GEx)',
                                            'gene' : 'SYMBOL'})

df_GExp_all.sort_values('log2_FC (GEx)', ascending = False)
```

## Plot visualization

### all

```python
#volcano plot
# Plot, after convertion  y to log10(pvalue)
plt.scatter(x=df_GExp_all['log2_FC (GEx)'],y=df_GExp_all['padj(GEx)'].apply(lambda x:-numpy.log10(x)),s=5)

plt.xlabel("Log2.FoldChange")
plt.ylabel("adjp.value -log10")

# pvalue <0.05
plt.axhline(-numpy.log10(0.05),color="grey",linestyle="--")
```

### Significant ones

```python
# keep padj < 0.05
df_GExp_signif = df_GExp_all[df_GExp_all['padj(GEx)'] < 0.05]
df_GExp_signif.shape
```

```python
df_GExp_signif.head()
```

```python
#volcano plot
# Plot, after convertion  y to log10(pvalue)
top_15 = df_GExp_signif.sort_values(by='padj(GEx)', ascending=True).head(15)
topneg_5 = df_GExp_signif.sort_values(by='log2_FC (GEx)', ascending=True).head(5)

plt.scatter(x=df_GExp_signif['log2_FC (GEx)'], 
            y=df_GExp_signif['padj(GEx)'].apply(lambda x:-numpy.log10(x)),s=5)

# Annotate points on the plot with Symbol names
for index, row in top_15.iterrows():
    plt.annotate(row['SYMBOL'], (row['log2_FC (GEx)'], -numpy.log10(row['padj(GEx)'])),
                fontsize = 10)
    
for index, row in topneg_5.iterrows():
    plt.annotate(row['SYMBOL'], (row['log2_FC (GEx)'], -numpy.log10(row['padj(GEx)'])),
                fontsize = 10)

plt.xlabel("Log2.FoldChange")
plt.ylabel("adjp.value -log10")

# pvalue <0.05
plt.axhline(-numpy.log10(0.05),color="grey",linestyle="--")
```

```python
# save to .txt the genes
## positive
numpy.savetxt(file_directory+"3) output/top15_genes_HEK293.txt", 
              top_15['SYMBOL'].unique(), delimiter="\t", fmt="% s")

## negative
numpy.savetxt(file_directory+"3) output/top5_negative_genes_HEK293.txt", 
              topneg_5['SYMBOL'].unique(), delimiter="\t", fmt="% s")

print(top_15.head(), '\n','\n',  topneg_5)
```
