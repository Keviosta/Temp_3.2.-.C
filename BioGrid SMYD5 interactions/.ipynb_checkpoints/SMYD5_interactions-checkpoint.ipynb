# Data import

### BioGrid SMYD5 interactions

```python
#data BioGrid SMYD5 interactions
df_bioGrid = pandas.read_csv(file_directory+'1) input/bioGrid/biogrid_smyd5_interaction_list.txt', sep='\t')
df_bioGrid = df_bioGrid.iloc[:,7:9] #only symbols
df_bioGrid = df_bioGrid[(df_bioGrid != 'tat').all(axis=1)] # drop HIV
df_bioGrid['node1'] = df_bioGrid['node1'].replace('PARK2', 'PRKN')
df_bioGrid['string'] = 'NO'
df_bioGrid.head(5)
```

```python
list_all_bioGrid = list(df_bioGrid.iloc[:,0]) + list(df_bioGrid.iloc[:,1])
list_all_bioGrid = set(list_all_bioGrid)
print(len(list_all_bioGrid))
```

### STRING from BioGrid SMYD5 interactions

```python
#data BioGrid SMYD5 interactions
df_STRING_formbioGrid = pandas.read_csv(file_directory+'1) input/bioGrid/string_interactions_from_biogrid_list.tsv', sep='\t')
df_STRING_formbioGrid['string'] = 'YES'
df_STRING_formbioGrid = df_STRING_formbioGrid[['node1', 'node2', 'string']]
df_STRING_formbioGrid
```

```python
list_all_string_frombioGrid = list(df_STRING_formbioGrid.iloc[:,0]) + list(df_STRING_formbioGrid.iloc[:,1])
list_all_string_frombioGrid = set(list_all_string_frombioGrid)
print(len(list_all_string_frombioGrid))
```

### STRING SMYD5 interactions

```python
#data BioGrid SMYD5 interactions
df_STRING_smyd5inter = pandas.read_csv(file_directory+'1) input/bioGrid/string_smyd5_interactions.tsv', sep='\t')
df_STRING_smyd5inter['string'] = 'YES'
df_STRING_smyd5inter = df_STRING_smyd5inter[['node1', 'node2', 'string']]
df_STRING_smyd5inter.head()
```

```python
list_all_string = list(df_STRING_smyd5inter.iloc[:,0]) + list(df_STRING_smyd5inter.iloc[:,1])
list_all_string = set(list_all_string)
print(len(list_all_string))
```

### Merge

```python
list_all = set(list(list_all_string) + list(list_all_string_frombioGrid) + list(list_all_bioGrid))
len(list_all)
```

```python
df_all = pandas.concat([df_bioGrid, df_STRING_formbioGrid, df_STRING_smyd5inter], axis=0)
df_all
```

```python
path = "/Users/kja11/OneDrive - Menntaský/PostDoc_Hypothermia/SMYD5 interactions databases/"
df_all.to_csv('{}/smyd5_interaction_to_cytoscape.csv'.format(path), sep=',')
```

### Add Slope information from proteomic for these proteins

```python
df_linregress = pandas.read_csv(file_directory+'3) output/Proteomics/df_linregress_noCTL.csv', 
                     sep = ',', index_col = 'Gene')
df_linregress.reset_index(inplace=True)
df_linregress.head()
```

```python
# select all the proteins from the previous nodes
df_nodes = pandas.concat([df_bioGrid, df_STRING_formbioGrid, df_STRING_smyd5inter], axis=0)
df_nodes = df_nodes.iloc[:,:2]
nodes = pandas.concat([df_nodes['node1'], df_nodes['node2']], ignore_index=True)
new_df = pandas.DataFrame(nodes, columns=['Gene'])
new_df.drop_duplicates(subset = "Gene", inplace = True)
print(len(new_df))
new_df.head()
```

```python
# Merge the two to have the slope information
df_merge = pandas.merge(new_df, df_linregress, on= 'Gene', how='left', indicator = True)
df_merge = df_merge.iloc[:,:2]
df_merge['slope'] = df_merge['slope'].fillna('NO')
df_merge
```

```python
merge_nono = df_merge[df_merge.slope != 'NO']
merge_no = df_merge[df_merge.slope == 'NO']
```

```python
path = "/Users/kja11/OneDrive - Menntaský/PostDoc_Hypothermia/SMYD5 interactions databases/"
df_merge.to_csv('{}/interaction_nodes_withSlope_to_cytoscape.csv'.format(path), sep=',')
merge_nono.to_csv('{}/interaction_nodes_withSlope_VALUE_to_cytoscape.csv'.format(path), sep=',')
merge_no.to_csv('{}/interaction_nodes_withSlope_NO_to_cytoscape.csv'.format(path), sep=',')
```
