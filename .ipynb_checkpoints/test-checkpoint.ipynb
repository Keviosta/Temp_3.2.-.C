```python
import matplotlib, matplotlib.pyplot as plt
import numpy as np
```

```python
matplotlib.rcParams.update({'svg.fonttype' : 'none'})
```

```python
file_directory = "/Users/kja11/OneDrive - Menntask√Ω/PostDoc_Hypothermia/in_silico/Python/"
```

```python
# Define the categories and values
categories = ['Anxiety', 'Meltdowns', 'Eye Contact', 'Social Issues', 'Sensory Processing', 'Executive Functioning', 'Emotional Regulation', 'Stimming']

# Create a figure with polar projection
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(polar=True)

angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()

# Make the plot cyclic
angles += angles[:1]

# Draw the plot
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)

title_file = 'circulaire'
plt.savefig(file_directory+'3) output/'+title_file+'.svg')
```
