# II] Data Download

## a. Expression File

```python
#data all_counts_filtered
df_count = pandas.read_csv(file_directory+path+ 'transcript_HEK293_temp_all_counts_unfiltered.tsv',
                     sep = '\t')
print(df_count.shape)
df_count.head()
```

```python
# Remove the non expressed transcripts (less than 10 counts)
full_leng = len(df_count)
df_count = df_count[df_count.max(axis=1) >= 10]

filtr_leng = len(df_count)
dropped_transcripts = full_leng - filtr_leng

print(f'On {full_leng}, {dropped_transcripts} transcripts do not exceed 10 counts')

df_count
```

```python
# calculate mean
df_count = df_count.assign(mean_32C = (df_count[['s37rep1', 's37rep2', 's37rep3']].mean(axis=1).round(0)),
                           mean_37C = (df_count[['s32rep1', 's32rep2', 's32rep3']].mean(axis=1).round(0))) 

df_count.rename(columns={df_count.columns[2]: 'transcript_id'},inplace=True)
df_count = df_count[['transcript_id','gene_name', 'mean_37C', 'mean_32C']]
print(len(df_count))
df_count.head()
```

## b. Transcript information file

```python
%%time
#data DETseq HEK293 
df_length = pandas.read_csv(file_directory+path+ '20240216_mytranscripts.gtf',
                               sep = '\t', comment='#', header=None)
print(df_length.shape)

df_length.columns = ["seqname", "source", "feature", "start", "end",
                  "score", "strand", "frame", "attribute"]
df_length.head()
```

```python
df_length.feature.value_counts()
```

```python
# coding sequence (CDS) regions
df_length = df_length[df_length["feature"] == "CDS"]

#Group CDS entries by transcript ID
df_length['transcript_id'] = df_length['attribute'].str.extract(r'transcript_id "([^"]+)"')

# Sum up CDS lengths per transcript
df_length['cds_length'] = df_length['end'] - df_length['start'] + 1

# df_length.set_index('transcript_id', inplace = True, drop = True )
df_length = df_length[['transcript_id', 'cds_length']]
print(df_length.describe())
df_length
```

```python
df_length = df_length.groupby('transcript_id')['cds_length'].sum()
df_length = df_length.reset_index()
df_length
```

## c. Annotated data

```python
#data DEseq HEK293 
df_annotated = pandas.read_csv(file_directory+'1) input/RNAseq/from_output/DETranscripts_HEK293_temp_annotated.csv',
                               sep = ',')

print(len(df_annotated))
df_annotated = df_annotated.dropna(axis=0)
print(len(df_annotated))
df_annotated.head(2)
```

## d. Merge

```python
df = df_length.merge(df_count)

# calculate ratio - expression/size
df = df.assign(ratio_expsize_37C = df['mean_37C'] / df['cds_length'],
               ratio_expsize_32C = df['mean_32C'] / df['cds_length'])

df = df.merge(df_annotated)
df = df[['Transcript name','transcript_id', 'cds_length', 'Transcript type',
         'mean_37C', 'mean_32C', 'ratio_expsize_37C', 'ratio_expsize_32C', 'log2FoldChange','padj']]
df.head(2)
```

## Subsets

```python
# select the significant one (differentially expressed)
df_signif = df[(df.padj < 0.05)]
print(f'{len(df_signif)} transcripts are significantly different between 37¬∞ and 32¬∞C')

df_signif.head(2)
```

```python
df_sign_up = df_signif[df_signif['log2FoldChange'] > 0]
df_sign_down = df_signif[df_signif['log2FoldChange'] < 0]
```

```python
df_signif = df_signif.assign(effect=numpy.where(df_signif['log2FoldChange'] > 0, 'Up regul.', 'Down regul.'))
df_signif.head()
```

```python
df_signif.effect.value_counts()
```

# Methods

```python
df_sign_up = df_signif[df_signif['log2FoldChange'] > 0]
df_sign_down = df_signif[df_signif['log2FoldChange'] < 0]

effects = df_sign_up, df_sign_down
colors = 'gold', 'blueviolet'

for i in range(len(effects)):
    effect = effects[i]
    color = colors[i]
    
    size = effect[['cds_length']].apply(numpy.log10)
    values = size.values.flatten()
    
    #plot
    seaborn.histplot(values,color=color, kde=True, alpha=0.2)
    legend_elements = [Patch(facecolor='gold', label = f'{len(df_sign_up)} Up regul. transcripts'),
                       Patch(facecolor='blueviolet', label = f'{len(df_sign_down)} Down regul. transcripts')]
    
    plt.legend(handles=legend_elements, loc='upper right', fontsize=12)
    plt.title('Distribution of the differentially\n expressed transcripts size at 32¬∞C')
    plt.xlabel('log10 total coding sequence (CDS) size (bp)')
    plt.ylabel('Count')
```
